```python
# Append Function
# -- In pandas append function is like adding one table (DF) to the bottom of another table.
# -- It combines data from two tables vertically, stacking one on the top of other.
# -- you have two tables with same columns one containing inf oabout students in a class & other'
# -- containing information about other class. If I want to combine both classes into a single 
# -- table then we had to use append method for that purpose
```

```python
# create the 1st DF
data1 = {'Name':['Alice','Bob'],'Age':[25,30]}
df1 = pd.DataFrame(data1)
df1
```

```python
#  Create the 2nd DF
data2 = {'Name':['Charlie','David'],'Age':[22,35]}
df2 = pd.DataFrame(data2)
df2
```

```python
# Append the second DF to the First_One
combined_df = pd.concat([df1,df2],ignore_index=True)
combined_df
```

```python
# Merge Method in Pandas
# Merge is use to combine the two DF -- base on the common column 
# Create the first DF
data1 = {'StudentID':[1,2,3],'Name':['Alice','Bob','Charlie']}
student_df = pd.DataFrame(data1)
student_df
```

```python
# Create the second DF
data2 = {'StudentID':[2,3,4],'Score':[85,92,78]}
scores_df = pd.DataFrame(data2)
scores_df
```

```python
# Merge the two DF based on the 'StudentID' column
merged_df = pd.merge(student_df,scores_df,on='StudentID')
merged_df
```

```python
df1 = pd.DataFrame({'A':[1,2,3]},index = ['a','b','c'])
df2 = pd.DataFrame({'B':[4,5,6]},index = ['a','b','d'])
# using the join method in the pandas
# joining the df1 with df2 on index
result = df1.join(df2,how='left')
result
```

```python
result = df1.join(df2,how='right')
result
```

```python
result = df1.join(df2,how='outer')
result
```

```python
# If you want too create an array with the defined data type in the numpy then you can create so
# use dtype to pass dtype
na = np.array([1,2,3,4],dtype='S') # here S indicate the string so S is use to define datatype
print(na)
print(na.dtype)
```

```python
# Change datatype from integer to the boolean
na = np.array([1,0,3])
na_copy = na.astype(bool)
print(na_copy)
print(na_copy.dtype)
```

```python
## Difference between the copy & the view of array
# -- the copy is the new array while a view is just the view of the original array
# -- any changes made in the copy array will not affect the original array but if you make any 
# -- changes to the view of array this will reflect in your original array
# -- view does not own the data and copy own its own data
```

```python
# Example of copy
na = np.array([1,2,3,4,5])
x = na.copy()
print(na)
print(x)
```

```python
na[0] = 42
```

```python
print(na)
print(x)
```

```python
print(na)
print(x)
```

```python
# Example of view -- change in the view will affect the original array
na = np.array([1,2,3,4,5])
x = na.view()
print(na)
print(x)
```

```python
print(na)
print(x)
```

```python
# Joining the numpy array's --> means putting the content of two or more array's in the single 
# array's --> In SQL we join tables by the key while in numpy we join the array by using the axis
# both are same
# we can use concat function for that purpose -- along with the axis -- by default axis is the 0
```

```python
# Joining the two_array's
na1 = np.array([1,2,3])
na2 = np.array([4,5,6])
na3 = np.concatenate((na1,na2))
na3
```

```python
# Concatnating / joining 2 2D arrays along with coulmns (axis=1)
na1 = np.array([[1,2],[3,4]])
na2 = np.array([[5,6],[7,8]])
na3 = np.concatenate((na1,na2),axis=1)
na3
```

```python
# Concatnating / joining 2 2D arrays along with rows (axis=0)
na1 = np.array([[1,2],[3,4]])
na2 = np.array([[5,6],[7,8]])
na3 = np.concatenate((na1,na2),axis=0)
na3
```

```python
# Joining the arrays using the stack function -- stack function is same as concatnate function
# the only difference is that stacking is done along the new axis
na1 = np.array([1,2,3])
na2 = np.array([4,5,6])
na3 = np.stack((na1,na2),axis=1)
print(na3)
```

```python
# Different types of stack -- there are two types of stack -- stack along row's and the stack 
# along column's
na1 = np.array([1,2,3])
na2 = np.array([4,5,6])
na3 = np.hstack((na1,na2)) # hstack used for stacking via row
na4 = np.vstack((na1,na2)) # vstack used for stacking via column's
na5 = np.dstack((na1,na2)) # dstack used for stacking along the height
print(na3)
print(na4)
print(na5)
```
